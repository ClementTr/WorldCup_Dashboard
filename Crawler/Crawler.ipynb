```python
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import time
```

```python
browser = webdriver.Firefox()
browser.maximize_window()
browser.get("https://pro.doctolib.fr/bo-accueil")

username = browser.find_element_by_name("username")
password = browser.find_element_by_name("password")
submit = browser.find_element_by_name("commit")

username.send_keys("tailleur.osteo@gmail.com")
password.send_keys("46884688")
submit.click()
time.sleep(1)

dropdown = browser.find_element_by_class_name("dropdown-toggle")
dropdown.click()

time.sleep(1)
patients_link = browser.find_element_by_xpath('//a[@href="/b#/patients"]')
patients_link.click()
time.sleep(2.5)

list_birthdays = []
list_firstnames = []
list_lastnames = []
list_mails = []
list_phones = []
list_post_codes = []

for compteur in range(1, 169):
    print(browser.current_url)
    patients = browser.find_elements_by_xpath(".//tbody/tr/td[2]")
    for patient in patients:
        patient.click()
        time.sleep(1.5)
        # Sex
        '''sex_input = browser.find_element_by_id("gender-0")
        if sex_input.get_attribute('value') == "false":
            list_sexs.append("Male")
        elif sex_input.get_attribute('value') == "true":
            list_sexs.append("Female")
        else:
            list_sexs.append("")'''
        # Post Code
        post_code = browser.find_element_by_xpath("//*[@title='Code postal']")
        list_post_codes.append(post_code.get_attribute('value'))
        # Birthday
        birthday_input = browser.find_element_by_id("birthdate")
        list_birthdays.append(birthday_input.get_attribute('value'))
        # Firstname
        firstname_input = browser.find_element_by_name("first_name")
        list_firstnames.append(firstname_input.get_attribute('value'))
        # Lastname
        lastname_input = browser.find_element_by_name("last_name")
        list_lastnames.append(lastname_input.get_attribute('value'))
        # Mail
        email_input = browser.find_element_by_name("email")
        list_mails.append(email_input.get_attribute('value'))
        # Phone
        phone_input = browser.find_element_by_id("phone_number")
        list_phones.append(phone_input.get_attribute('value'))

        # EXIT !
        exit_button = browser.find_element_by_class_name("dc-modal-close-button")
        exit_button.click()
        time.sleep(1.5)

    # Get Next Page
    pages = browser.find_elements_by_xpath('//a[@class="dl-link"]')
    for p in pages:
        if p.text == '›':
            p.click()
            time.sleep(3)
            break;
    

```

```python
df = pd.DataFrame({
    "Firstname": list_firstnames,
    "Lastname": list_lastnames,
    "Post Code": list_post_codes,
    "Mail": list_mails,
    "Birth": list_birthdays,
    "Phone": list_phones
})

df
```

```python
df.to_csv("data_doctolib_init.csv", index=False)
```

```python
df["Lastname"] = df["Lastname"].str.title()
df["Firstname"] = df["Firstname"].str.title()
df.head()
```

```python
df_cleaned = df[(df.Lastname != "") & (df.Lastname != "X")].copy()
df_cleaned.to_excel("data_doctolib_cleaned.xlsx", index=False)
df_cleaned.head()
```
